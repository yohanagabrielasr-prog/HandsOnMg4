9. Diketahui dari suatu tabel mortalita, jumlah penduduk yang hidup
pada usia tepat 20 tahun (𝑙20) adalah 98.567 jiwa, dan yang hidup
pada usia tepat 21 tahun (𝑙21) adalah 98.432 jiwa. Hitunglah jumlah
penduduk yang diprediksi hidup pada usia 20 tahun dan 8 bulan
(𝑙20,6667) dengan menggunakan asumsi Uniform Distribution of
Deaths (UDD).

```python
# ============================================================
# PERHITUNGAN AKTUARIA
# ============================================================
# Topik:
# 1. UDD (Uniform Distribution of Deaths)
# 2. CFM (Constant Force of Mortality)
# ============================================================

import math

print("===================================================")
print("Hands On Minggu 4 Matematika Aktuaria")
print("===================================================\n")

# ============================================================
# SOAL 9
# ============================================================
print("SOAL 9 - UDD")
print("-----------------------------------------------")

# Diketahui:
# l20 = 98.567
# l21 = 98.432

l20 = 98567
l21 = 98432

# Rumus dasar:
# d_x = l_x - l_(x+1)
# l_(x+s) = l_x - s * d_x   (Asumsi UDD)

# Hitung d20
d20 = l20 - l21

# 8 bulan = 8/12 tahun
s = 8/12

# Hitung l_(20 + 8/12)
l20_8bulan = l20 - s * d20

print(f"d20 = l20 - l21 = {d20}")
print(f"s = 8/12 = {s}")
print(f"l_(20 + 8/12) = l20 - s*d20")
print(f"= {l20} - ({s} × {d20})")
print(f"= {l20_8bulan}\n")


```

10. Sebuah perusahaan asuransi memiliki data bahwa probabilitas
kematian antara usia 40 tahun hingga 41 tahun (𝑞40) adalah
0,0025. Hitunglah probabilitas seseorang yang berusia tepat 40
tahun 3 bulan (𝑥 + 𝑠 = 40,25) akan meninggal dalam satu tahun ke
depan (𝑞40,25) dengan asumsi UDD.

```python
# ============================================================
# SOAL 10
# ============================================================
print("SOAL 10 - UDD")
print("-----------------------------------------------")

# Diketahui:
# q40 = 0.0025
q40 = 0.0025

# 3 bulan = 3/12 tahun
s = 3/12

# Rumus UDD untuk usia pecahan:
#
# q_(x+s) = ((1 - s) * q_x) / (1 - s*q_x)

numerator = (1 - s) * q40
denominator = 1 - s * q40
q40_025 = numerator / denominator

print("Rumus:")
print("q_(x+s) = ((1 - s) * q_x) / (1 - s*q_x)\n")

print(f"Pembilang = (1 - {s}) × {q40} = {numerator}")
print(f"Penyebut  = 1 - ({s} × {q40}) = {denominator}")
print(f"q_(40.25) = {q40_025}\n")
```

11. Diketahui bahwa force of mortality (𝜇𝑥) adalah konstan pada
interval usia 50-51 tahun sebesar 0,005. Hitunglah probabilitas
seseorang yang berusia tepat 50 tahun 6 bulan (𝑥 + 𝑠 = 50,5) akan
bertahan hidup hingga usia 51 tahun 6 bulan (𝑝50,5) berdasarkan
asumsi CFM.

```python
# ============================================================
# SOAL 11
# ============================================================
print("SOAL 11 - CFM")
print("-----------------------------------------------")

# Diketahui:
# μ = 0.005 (konstan)
mu = 0.005

# Interval waktu 1 tahun
t = 1

# Rumus CFM:
#
# p = e^(-μt)

p50_5 = math.exp(-mu * t)

print("Rumus:")
print("p = e^(-μt)\n")

print(f"p_(50.5) = e^(-{mu} × {t})")
print(f"= {p50_5}\n")
```

12. Dengan data yang sama (𝜇𝑥 = 0,005 pada interval usia 50-51
tahun), hitunglah probabilitas seseorang yang berusia tepat 50
tahun 6 bulan (𝑥 + 𝑠 = 50,5) akan meninggal dalam satu tahun ke
depan (𝑞50,5) sebelum mencapai usia 51 tahun 6 bulan.

```python

# ============================================================
# SOAL 12
# ============================================================
print("SOAL 12 - CFM")
print("-----------------------------------------------")

# Rumus:
# q = 1 - p

q50_5 = 1 - p50_5

print("Rumus:")
print("q = 1 - p\n")

print(f"q_(50.5) = 1 - {p50_5}")
print(f"= {q50_5}\n")


print("===================================================")
print("SELESAI")
print("===================================================")
```
