1. Diketahui 𝑙0 = 1.000.000 dan 𝑙25 = 985.000.
Hitunglah nilai Survival Function 𝑆25, yaitu
probabilitas seorang bayi baru lahir akan hidup
hingga usia 25 tahun.

```python
import math
# ============================================================
# 1. Survival Function S25
# Rumus:
# Sx = lx / l0
# ============================================================

l0 = 1_000_000
l25 = 985_000

S25 = l25 / l0

print("1. Survival Function")
print("S25 =", S25)
print("Probabilitas bayi hidup sampai usia 25 =", S25*100, "%\n")
```

2. Diketahui 𝑙40 = 98.567 dan 𝑞40 = 0,0025.

3. Diketahui dalam tabel ultimate, 𝑞50 = 0,0018.

berusia 50 tahun akan hidup mencapai usia 51
tahun.

4. Menurut Hukum Gompertz, force of mortality
dinyatakan
sebagai 𝜇𝑥 = 𝐵𝑐𝑥.
diketahui 𝐵 = 0,00004 dan 𝑐 = 1,09,
hitunglah 𝜇60.

```python
# ============================================================
# 4. Hukum Gompertz
# Rumus:
# μx = B * c^x
# ============================================================

B = 0.00004
c = 1.09
x = 60

mu60 = B * (c ** x)

print("4. Force of mortality Gompertz")
print("μ60 =", mu60, "\n")
```

5. Menurut Hukum Makeham, force of mortality
dinyatakan
sebagai 𝜇𝑥 = 𝐴+𝐵𝑐^𝑥.
diketahui 𝐴 = 0,001, 𝐵 = 0,00003, jika c =
1,08, hitunglah 𝜇65.



```python
# ============================================================
# 5. Hukum Makeham
# Rumus:
# μx = A + B*c^x
# ============================================================

A = 0.001
B = 0.00003
c = 1.08
x = 65

mu65 = A + B * (c ** x)

print("5. Force of mortality Makeham")
print("μ65 =", mu65, "\n")
```

6. Sebuah perusahaan asuransi menggunakan
tabel seleksi dengan periode seleksi 2 tahun.
Diketahui:

[40] = 0,0009

[40]+1 = 0,0012

*q*42 = 0,0020 (ultimate)

Jika 𝑙[40] = 100.000, hitunglah:

a. 𝑙[40]+1 dan 𝑙[40]+2

b. 𝑙42 (ultimate)

```python
# ============================================================
# 6. Tabel Seleksi (periode 2 tahun)
# ============================================================

l_sel40 = 100000
q_sel40 = 0.0009
q_sel40_1 = 0.0012

# l[40]+1
l_sel40_1 = l_sel40 * (1 - q_sel40)

# l[40]+2
l_sel40_2 = l_sel40_1 * (1 - q_sel40_1)

print("6. Tabel Seleksi usia 40")
print("l[40]+1 =", l_sel40_1)
print("l[40]+2 =", l_sel40_2, "\n")
```

12. Sebuah perusahaan asuransi menggunakan
periode seleksi 2 tahun. Diketahui:

[60] = 10.000

[60] = 0,004

[60]+1 = 0,005

62 = 0,007 (ultimate)

Hitunglah:
a. 𝑙[60]+1 dan 𝑙[60]+2
b. Probabilitas seseorang yang terpilih pada
usia 60 akan hidup hingga usia 63 (𝑝^3[60])

```python
# ============================================================
# 12. Tabel seleksi usia 60
# ============================================================

l_sel60 = 10000
q_sel60 = 0.004
q_sel60_1 = 0.005
q62 = 0.007

# l[60]+1
l_sel60_1 = l_sel60 * (1 - q_sel60)

# l[60]+2
l_sel60_2 = l_sel60_1 * (1 - q_sel60_1)

# Probabilitas hidup sampai 63
p3_60 = (1 - q_sel60) * (1 - q_sel60_1) * (1 - q62)

print("12. Tabel Seleksi usia 60")
print("l[60]+1 =", l_sel60_1)
print("l[60]+2 =", l_sel60_2)
print("Probabilitas hidup sampai usia 63 =", p3_60)
```
